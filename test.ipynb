```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)  # Semilla para reproducibilidad
data_size = 100

# Datos simulados: dos columnas de valores aleatorios
x_values = np.linspace(0, 10, data_size)
y_values = 3 * x_values + 7 + np.random.normal(0, 3, data_size)  # A√±adimos ruido gaussiano

# Crear un DataFrame
data = pd.DataFrame({
    "X": x_values,
    "Y": y_values
})

# Mostrar las primeras filas del DataFrame
print("\nPrimeras filas del DataFrame:")
print(data.head())

# Mostrar estad√≠sticas descriptivas
print("\nEstad√≠sticas descriptivas del DataFrame:")
print(data.describe())

# Crear una gr√°fica
plt.figure(figsize=(10, 6))

# Gr√°fico de dispersi√≥n
plt.scatter(data["X"], data["Y"], color='skyblue', label="Datos Aleatorios", alpha=0.7)

# L√≠nea de tendencia ideal
plt.plot(data["X"], 3 * data["X"] + 7, color='orange', label="Tendencia Ideal", linestyle='--')

# A√±adir l√≠nea de regresi√≥n ajustada
coeffs = np.polyfit(data["X"], data["Y"], 1)
polynomial = np.poly1d(coeffs)
plt.plot(data["X"], polynomial(data["X"]), color='green', label="Regresi√≥n Ajustada", linewidth=2)

# Personalizar la gr√°fica
plt.title("Gr√°fico de Datos Aleatorios con Regresi√≥n Ideal y Ajustada")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```
